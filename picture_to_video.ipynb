```python
# 文字轉語音
import pyttsx3
from pydub import AudioSegment

# the engine
engine = pyttsx3.init()

# 'Hello World' is just an example
engine.save_to_file('有一天，科技的進步給了世界一個未來的新交通工具: 飛船', 'picture1.mp3')
engine.save_to_file('飛船可以讓人們從一個地方到另一個地方的時間極短，甚至可以在極短時間內完成地球公轉。飛船可以把人們從一個地方帶到另一個地方，而旅行只需要幾分鐘。飛船可以航行在大氣層之外，不受地球重力的影響，可以提供更舒適安全的旅行環境。而且，它可以停在任何地方，讓乘客可以安全出入。', 'picture2.mp3')
engine.save_to_file('這種新的交通工具讓旅行更加方便和快捷，可以讓人們快速到達他們想去的地方，還可以跨越高山、低谷、海洋等不同的地域。', 'picture3.mp3')
engine.runAndWait()

# picture1.mp3 的長度
audio = AudioSegment.from_file("picture1.mp3")
picture1_duration = audio.duration_seconds
print(f'{picture1_duration}秒')

# picture2.mp3 的長度
audio = AudioSegment.from_file("picture2.mp3")
picture2_duration = audio.duration_seconds
print(f'{picture2_duration}秒')

# picture3.mp3 的長度
audio = AudioSegment.from_file("picture3.mp3")
picture3_duration = audio.duration_seconds
print(f'{picture3_duration}秒')
```

```python
# 單張找片轉影片
# 第一張照片
img_clip = ImageClip("picture1.png", transparent=True).set_duration(picture1_duration)  #看那段話語音念多久，照片就呈現多長
img_clip.write_videofile("picture1.mp4",fps=30, temp_audiofile="temp-audio.m4a", remove_temp=True, codec="libx264", audio_codec="aac")
print('ok')

# 第二張照片
img_clip = ImageClip("picture2.png", transparent=True).set_duration(picture2_duration)
img_clip.write_videofile("picture2.mp4",fps=30, temp_audiofile="temp-audio.m4a", remove_temp=True, codec="libx264", audio_codec="aac")
print('ok')

# 第三張照片
img_clip = ImageClip("picture3.png", transparent=True).set_duration(picture3_duration)
img_clip.write_videofile("picture3.mp4",fps=30, temp_audiofile="temp-audio.m4a", remove_temp=True, codec="libx264", audio_codec="aac")
print('ok')

```

```python
# 讀入音軌
audio1 = AudioFileClip("picture1.mp3")
audio2 = AudioFileClip("picture2.mp3")
audio3 = AudioFileClip("picture3.mp3")
# 讀入視頻
video1 = VideoFileClip("picture1.mp4").set_audio(audio1)
video2 = VideoFileClip("picture2.mp4").set_audio(audio2)
video3 = VideoFileClip("picture3.mp4").set_audio(audio3)
# 合併視頻
output = concatenate_videoclips([video1.crossfadeout(2), video2.crossfadein(2).crossfadeout(2), video3.crossfadein(2)],method="compose")
# 預覽
output.ipython_display()
# 輸出
video1.write_videofile("output.mp4",fps=30,audio=True, temp_audiofile="temp-audio.m4a", remove_temp=True, codec="libx264", audio_codec="aac")


```
